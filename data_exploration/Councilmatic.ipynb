# Person Votes

```python
sql = """
select voter_name, option, count(*) as cnt
from opencivicdata_personvote
group by voter_name, option
order by voter_name, option
"""
person_votes = pd.read_sql_query(sql, db.engine)
person_votes
```

```python
councilmembers = person_votes.voter_name.unique()
councilmembers
```

```python
for councilmember in councilmembers:
    curr_person_votes = person_votes[person_votes['voter_name'] == councilmember]
    
    y = curr_person_votes['option']
    y_pos = np.arange(len(y))
    x = curr_person_votes['cnt']
    
    plt.rcdefaults()
    fig, ax = plt.subplots()
    
    ax.barh(y_pos, x, align='center', color='blue', ecolor='black')
    ax.set_yticks(y_pos)
    ax.set_yticklabels(y)
    ax.invert_yaxis()  # labels read top-to-bottom
    ax.set_xlabel('Number of Votes')
    ax.set_title(councilmember)
    
    plt.show()
```
